```python
card_use_pattern['gender'] = card_use_pattern['gender'].apply(lambda x : 'ETC' if x not in ['F', 'M'] else x)
temp = card_use_pattern[card_use_pattern['gender'] != 'ETC'].groupby(['gender'])['age'].value_counts().unstack().sort_index()
temp.rename(index={'F' : 'Female', 'M' : 'Male'}, inplace = True)
man = temp.loc['Male']
woman = -temp.loc['Male']

fig, ax = plt.subplots(1, 1, figsize = (16, 8))
ax.bar(man.index, man, width = 0.4, color = 'salmon', alpha = 0.5, label = 'Male')
ax.bar(woman.index, woman, width = 0.4, color = 'skyblue', alpha = 0.5, label = 'Female')
plt.rcParams.update({'font.size' : 20})
for i in man.index:
    ax.annotate(f"{man[i]}",
               xy = (i, man[i] + 3),
               va = 'center', ha = 'center', fontweight = 'light', fontfamily = 'serif', fontsize = 13)

    
for i in woman.index:
    ax.annotate(f"{-woman[i]}",
               xy = (i, woman[i] - 3),
               va = 'center', ha = 'center', fontweight = 'light', fontfamily = 'serif', fontsize = 13)

for s in ['top', 'left', 'right', 'bottom']:
    ax.spines[s].set_visible(False)
    
fig.text(0.16, 0.95, '카드 사용자 고객의 성별/나이 분포', fontsize=30, fontweight='bold')
ax.set_yticks([])
plt.show()
```

```python
temp = card_use_code['upjoing_cd']
temp2 = card_use_code['class1']
card_joing = dict()
for i in range(len(temp)):
    temp[i] = temp[i].replace('s', 'S')
    card_joing[temp[i]] = temp2[i]
card_use_code['upjoing_cd'] = temp
card_use_pattern['class'] = card_use_pattern['upjoing_cd'].map(card_joing)

x = card_use_pattern['class'].value_counts().index
y = card_use_pattern['class'].value_counts().values
plt.rcParams.update({'font.size' : 15})
plt.figure(figsize = (20, 10))
plt.barh(x,y, color = 'skyblue', alpha = 0.5)
plt.title('업종별 카드 이용 현황', fontsize = 30)
plt.xlabel('카드 이용 현황', fontsize = 20)
plt.yticks(fontsize = 20)
plt.ylabel('업종별')
plt.show()
```
